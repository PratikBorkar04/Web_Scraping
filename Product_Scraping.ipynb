```python
#importing required modules
from flask import Flask, render_template, request,jsonify
import requests
from bs4 import BeautifulSoup as bs
from urllib.request import urlopen as uReq
import pandas as pd
from tkinter import *
import tkinter.ttk as ttk
import csv
import logging

#Product_scrapper class
class web_test:
    
    #logging configuration
    logging.basicConfig(filename = "product_Sc.log",level = logging.DEBUG , format = "%(asctime)s %(message)s")
    
    #input function
    def __init__(self,a,b):
        logging.info('step 1 --- Input from user')
        self.a = a
        self.b = b
        
    #web scrapping function
    def web_scrapping(self):
        
        #Clearing old data
        logging.info('Step 2 --- Clearing data and Adding header')
        file_name = self.a+'.csv'
        f = open(file_name, "w+")
        f.close()
        
        #Process of Adding header to csv file
        data = {'storage_color':[],
                    'Cost':[],
                   'Camera':[],
                    'Rating':[]}

        df = pd.DataFrame(data)
        df.to_csv(file_name, mode='a', index=False, header=True)
            
        #Loop for iterations / number of reviews required
        for i in range(1,(self.b)+1):
            
            logging.info('Step 3 --- inside Scrapper')
            
            flipkart_url = "https://www.flipkart.com/search?q=" + self.a
            print(flipkart_url)
            print("\t")
            uClient = uReq(flipkart_url)
            flipkartPage = uClient.read()
            flipkart_html = bs(flipkartPage, "html.parser")

            bigboxes = flipkart_html.findAll("div", {"class": "_1AtVbE col-12-12"})
            box = bigboxes[0]
            del bigboxes[0:3]

            box = bigboxes[i]
            productLink = "https://www.flipkart.com" + box.div.div.div.a['href']
            print(productLink)
            prodRes = requests.get(productLink)

            prodRes.encoding='utf-8'
            prod_html = bs(prodRes.text, "html.parser")
            print('\t')
            
            logging.info('Step 4 ---- Functions of product')
            #Color of the Product
            storage_color =  prod_html.find_all('span', {'class': "B_NuCI"})
            print(storage_color[0].text) 

            #Cost of the Product
            Cost = prod_html.find_all('div', {'class': "_30jeq3 _16Jk6d"})
            print(Cost[0].text)

            #Camera quality of the Product
            Camera = prod_html.find_all('div', {'class': "_3k-BhJ"})
            print(Camera[4].find_all('li', {'class': '_21lJbe'})[1].text)

            #Rating of the product
            Rating = prod_html.find_all('div', {'class': "_2d4LTz"})
            print(Rating[0].text)

            print("------------------------------------------------------------------------------------------------")
            print("\n")
    
            logging.info('Step 5 ---- Adding data to csv file')
            #Creating DataFrame
            data = {'storage_color':[storage_color[0].text],
                    'Cost':[Cost[0].text],
                   'Camera':[Camera[4].find_all('li', {'class': '_21lJbe'})[1].text],
                    'Rating':[Rating[0].text]}
            
            #Append DataFrame to csv file
            df = pd.DataFrame(data)
            df.to_csv(file_name, mode='a', index=False, header=False)
            
    def tk_view(self):
            logging.info('Step 6 ---- Inside Tkinter')
            root = Tk()
            #Title of the GUI
            root.title("Product_Scrapping")
            width = 500
            height = 400
            
            #GUI width and height
            screen_width = root.winfo_screenwidth()
            screen_height = root.winfo_screenheight()
            
            x = (screen_width/2) - (width/2)
            y = (screen_height/2) - (height/2)
            
            root.geometry("%dx%d+%d+%d" % (width, height, x, y))
            root.resizable(0, 0)
            
            # Tablemargin
            TableMargin = Frame(root, width=500)
            TableMargin.pack(side=TOP)
            
            # Horizontal and vertical Scrollbar 
            scrollbarx = Scrollbar(TableMargin, orient=HORIZONTAL)
            scrollbary = Scrollbar(TableMargin, orient=VERTICAL)
            tree = ttk.Treeview(TableMargin, columns=("storage_color","Cost","Camera","Rating"), height=400, selectmode="extended", yscrollcommand=scrollbary.set, xscrollcommand=scrollbarx.set)
            scrollbary.config(command=tree.yview)
            scrollbary.pack(side=RIGHT, fill=Y)
            scrollbarx.config(command=tree.xview)
            scrollbarx.pack(side=BOTTOM, fill=X)

            #Heading of the data
            tree.heading('storage_color', text="storage_color", anchor=W)
            tree.heading('Cost', text="Cost", anchor=W)
            tree.heading('Camera', text="Camera", anchor=W)
            tree.heading('Rating', text="Rating", anchor=W)

            tree.column('#0', stretch=NO, minwidth=0, width=0)
            tree.column('#1', stretch=NO, minwidth=0, width=200)
            tree.column('#2', stretch=NO, minwidth=0, width=200)
            tree.column('#3', stretch=NO, minwidth=0, width=300)
            tree.pack()

            logging.info('Step 7 ---- Fetching data from csv file')
            
            #Fetching data from csv file
            file = self.a +".csv"            
            with open(file) as f:
                reader = csv.DictReader(f, delimiter=',')
                for row in reader:
                    storage_color = row['storage_color']
                    Cost = row['Cost']
                    Camera = row['Camera']
                    Rating = row['Rating']

                    tree.insert("", 0, values=(storage_color,Cost,Camera,Rating))

            if __name__ == '__main__':
                root.mainloop()
```

```python
q = web_test('iphone12',2)
q.web_scrapping()
q.tk_view()
```
